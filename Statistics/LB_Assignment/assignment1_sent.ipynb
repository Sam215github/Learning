Write a program in python for calculating the following stats:

1. Interval Estimate 
2. Point Estimate
3. Hypothesis Testing  (Z and T)

-------------------------------------------------------------------------------

Result:
1. Hyothesis Test Result : This is Z-distribution and Reject Null Hypothesis
2. Interval Estimate is XXX to XXXX
3. Point Estimate is XXX

Note: Make Proper exception handling



```python
import csv

#Hardcoded aplha values for 0.025 and 0.05 
alpha_values = {'z-dist': {0.025: 1.96 , 0.05: 1.68}  , 't-dist': {0.025: 1.66 , 0.05: 0}} 

# To fetch interger and floating point values and value should be greater than Zero
def get_proper_val(str_val):
    
    value_str = input(str_val)
    
    if(value_str!= ''  ):
        value = eval(value_str)
        
        if isinstance(value, int) or isinstance(value, float) :
            if value <= 0:
                raise ValueError("Enter positive number!")                                                          
            return value
        else :
            print ('Enter Integer/Float value')
    else :
        raise Exception('Enter value')

#Hypothesis Test
def hypothesis_test(tail_flag,standard_deviation,sample_size,sample_mean,population_mean,distribution_type,alpha) :
    try :
        score_denominator = standard_deviation / (sample_size ** 0.5)

        if score_denominator!= 0 :
            score = (sample_mean - population_mean)/ (score_denominator)

            alpha_value = alpha_values[distribution_type][alpha]

            hypothesis_result = f'Hyothesis Test Result : This is {distribution_type} and '

            #prepare hypothesis result string 
            if score >= alpha_value :
                hypothesis_result += 'Reject Null Hypothesis'
            else :
                hypothesis_result += 'Accept Null Hypothesis'
                
    except Exception as e:
        print(f'Exception during {distribution_type} score calculation!!!')
        print(f'{e} Error occured!!!')
    return hypothesis_result+'\n';


#interval estimate calculation
def interval_estimate_calc(population_size,sample_size,distribution_type,alpha):
    if population_size > 0 : 
        proportion = sample_size/population_size
        alpha_value = alpha_values[distribution_type][alpha]
        margin_of_error = alpha_value * ( (proportion *(1-proportion) / population_size) ** 0.5)
    return f'Interval Estimate is {proportion-margin_of_error} to {proportion+margin_of_error}\n'

#common method for all types of test
def stats_calc(test_flag=4,tail_flag=1,population_mean=None, sample_mean=None,standard_deviation=None, sample_size=None,population_size= None, alpha =None):
       
    #check distribution  type     
    if sample_size >=30 :
        distribution_type = 'z-dist'
    else:
        distribution_type = 't-dist'

    #calculating alpha value based on type of tail test
    if tail_flag ==2 :
        alpha = alpha / 2
    
    stats_calc_result = ''
    #print(':::::::::::::::::::::::::::Result:::::::::::::::::::::::::')
    if test_flag ==1 or test_flag==4:
        stats_calc_result = interval_estimate_calc(population_size,sample_size,distribution_type,alpha)
    if test_flag == 2 or test_flag==4:
        point_estimate = population_mean;
        stats_calc_result += f'Point Estimate : {point_estimate}\n'
    if test_flag == 3 or test_flag==4:
        if sample_size > 0 :
            stats_calc_result += hypothesis_test(tail_flag,standard_deviation,sample_size,sample_mean,population_mean,distribution_type,alpha)
        else:
            print ('sample size should not be zero or less!!!')

    return stats_calc_result
    #print('------------------------------------------------------------------------------------')

#manual input     
def perform_manual_operation():
    input_type_test = get_proper_val('''Enter type of test
                                        1. Interval Estimate  
                                        2. Point Estimate 
                                        3. Hypothesis Test 
                                        4. All 
                                    ''')
    if input_type_test ==1:
        print('Give input for Interval Estimate :')
    elif input_type_test ==2:
        print('Give input for Point Estimate  :')
    elif input_type_test ==3:
        print('Give input for Hypothesis Test :')
    elif input_type_test ==4:
        print('Give input for All Test:')
    
    tail_flag=0
    sample_mean=0
    population_mean=0
    population_size=0
    sample_size=0
    standard_deviation=0
    alpha=0
    tail_flag=0
    
    if(input_type_test ==2 or input_type_test ==4 ):
        population_mean=get_proper_val('Enter Population mean value : ')
    if(input_type_test==3 or input_type_test ==4):
        sample_mean = get_proper_val('Enter Sample mean value : ')
      
    if(input_type_test != 2):
        population_size = get_proper_val('Enter population size : ')
        sample_size = get_proper_val('Enter Sample Size : ')
        standard_deviation = get_proper_val('Enter Standard deviation value : ')
        alpha = get_proper_val('Enter alpha value : ')
        tail_flag = get_proper_val('''Enter tail test ?  
                                        1. One-tail Test   
                                        2. Two-tail test
                                        ''')

    #call stats calculate method
    print(stats_calc(input_type_test,tail_flag,population_mean,sample_mean,standard_deviation,sample_size,population_size,alpha))
    
#csv file operation
#Not added many checks for input checking
def perform_file_operation():
    
    try :
        
        #you can give your input file path
        file_full_path = input('Enter CSV file full path : ')
        if file_full_path == '':
            file_full_path = 'C:\\Data\\git\\Learning\\Statistics\\LB_Assignment\\assignment_1_input.csv'
        

        with open(file_full_path,newline='\n') as csv_file:
            csv_reader = csv.reader(csv_file, delimiter=',')
            test_record = 0
            for row in csv_reader:
                
                tail_flag=0
                sample_mean=0
                population_mean=0
                population_size=0
                sample_size=0
                standard_deviation=0
                alpha=0
                tail_flag=0

                if test_record==0:
                    pass
                else:
                    if row[0] != '':
                        input_type_test= eval(row[0])
                    if row[1] != '':
                        population_mean= eval(row[1])
                    if row[2] != '':
                        sample_mean= eval(row[2])
                    if row[3] != '':
                        population_size= eval(row[3])
                    if row[4] != '':
                        sample_size= eval(row[4])
                    if row[5] != '':
                        standard_deviation= eval(row[5])
                    if row[6] != '':
                        alpha= eval(row[6])
                    if row[7] != '':
                        tail_flag= eval(row[7])
                    
                    print(f'CSV Record {test_record} result ::')
                    print(stats_calc(input_type_test,tail_flag,population_mean,sample_mean,standard_deviation,sample_size,population_size,alpha))
                    
                test_record+=1
    except Exception as e:
        print(f'Exception during csv operation!!!')
        print(f'Exception details: {e}!!!')

print('------------------------------------------------------------------------------------')
type_of_input = get_proper_val("""What do you want to provide 
                                          1. manual input
                                          2. csv file
                                          """)

if type_of_input == 2 :
    print('----------------------------Start CSV Operation--------------------------------------')
    perform_file_operation()
    print('----------------------------End CSV Operation----------------------------------------')
elif type_of_input != 2:
    print('----------------------------Start Manual Operation-------------------------------------')
    perform_manual_operation()
    print('----------------------------End Manual Operation--------------------------------------')
print('------------------------------------------------------------------------------------')
```
